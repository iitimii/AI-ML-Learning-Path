```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('darkgrid')
%matplotlib inline
```

```python
# Line Chart

years = range(2000, 2007)
yield_apples =np.linspace(0.5, 1, len(years))
yield_oranges = np.linspace(0.2, 1.4, len(years))
```

```python
plt.plot(years ,yield_apples, 's-b')
plt.plot(years, yield_oranges, 'o--r')
plt.ylabel('Yield (tons per hectare)')
plt.xlabel('Year');
plt.legend(['Apples', 'Oranges']);
```

```python
# Scatterplot
years = range(2000, 2030)
yield_apples =np.linspace(0.5, 4, len(years))
yield_oranges = np.linspace(0.2, 5, len(years))
yield_apples = np.random.rand(len(years)) + yield_apples
yield_oranges = np.random.rand(len(years)) + yield_oranges
```

```python
sns.scatterplot(x=yield_apples, y=yield_oranges, hue=years)
plt.ylabel('Oranges Yield (tons per hectare)')
plt.xlabel('Apples Yield (tons per hectare)');
plt.title('Apples yield against Oranges yield')
```

```python
# Bar Plot
total_yield = yield_oranges + yield_apples
plt.bar(years, total_yield)
plt.title("Total Yield")
plt.ylabel('Yield (tons per hectare)')
plt.xlabel('Year');
```

```python
# Pie Chart
yield_oranges = 23
yield_apples = 17
yield_bananas = 32
yield_grapes = 6
labels = ['Oranges', 'Apples', 'Bananas', 'Grapes']
plt.title("Percentage of fruits yield")
plt.pie([yield_oranges, yield_apples, yield_bananas, yield_grapes], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, explode=(0, 0.1, 0, 0));
```

```python
# Sub Plots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
years = range(2000, 2007)
yield_apples =np.linspace(0.5, 1, len(years))
yield_oranges = np.linspace(0.2, 1.4, len(years))
axes[0,0].plot(years ,yield_apples, 's-b')
axes[0,0].plot(years, yield_oranges, 'o--r')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].set_xlabel('Year');
axes[0,0].legend(['Apples', 'Oranges']);


years = range(2000, 2030)
yield_apples =np.linspace(0.5, 4, len(years))
yield_oranges = np.linspace(0.2, 5, len(years))
yield_apples = np.random.rand(len(years)) + yield_apples
yield_oranges = np.random.rand(len(years)) + yield_oranges

axes[0,1].set_ylabel('Oranges Yield (tons per hectare)')
axes[0,1].set_xlabel('Apples Yield (tons per hectare)');
axes[0,1].set_title('Apples yield against Oranges yield')
sns.scatterplot(x=yield_apples, y=yield_oranges, hue=years, ax=axes[0,1])


total_yield = yield_oranges + yield_apples
axes[1,0].bar(years, total_yield)
axes[1,0].set_title("Total Yield")
axes[1,0].set_ylabel('Yield (tons per hectare)')
axes[1,0].set_xlabel('Year');


yield_oranges = 23
yield_apples = 17
yield_bananas = 32
yield_grapes = 6
labels = ['Oranges', 'Apples', 'Bananas', 'Grapes']
axes[1,1].set_title("Percentage of fruits yield")
axes[1,1].pie([yield_oranges, yield_apples, yield_bananas, yield_grapes], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, explode=(0, 0.1, 0, 0));
```
