```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('darkgrid')
%matplotlib inline
```

```python
# Line Chart

years = range(2000, 2007)
yield_apples =np.linspace(0.5, 1, len(years))
yield_oranges = np.linspace(0.2, 1.4, len(years))
```

```python
plt.plot(years ,yield_apples, 's-b')
plt.plot(years, yield_oranges, 'o--r')
plt.ylabel('Yield (tons per hectare)')
plt.xlabel('Year');
plt.legend(['Apples', 'Oranges']);
```

```python
# Scatterplot
years = range(2000, 2030)
yield_apples =np.linspace(0.5, 4, len(years))
yield_oranges = np.linspace(0.2, 5, len(years))
yield_apples = np.random.rand(len(years)) + yield_apples
yield_oranges = np.random.rand(len(years)) + yield_oranges
```

```python
sns.scatterplot(x=yield_apples, y=yield_oranges, hue=years)
plt.ylabel('Oranges Yield (tons per hectare)')
plt.xlabel('Apples Yield (tons per hectare)');
plt.title('Apples yield against Oranges yield')
```

```python
# Bar Plot
total_yield = yield_oranges + yield_apples
plt.bar(years, total_yield);
```
